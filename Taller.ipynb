```python
#Punto 5
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp


# Definimos g(z) de manera segmentada usando sympy
g_piecewise = sp.Piecewise((3 * z - 1, z < 2), (7 - z, z >= 2))

# Definimos g(z) como una funci√≥n para evaluar la gr√°fica
def g_function(z_value):
    if z_value < 2:
        return 3 * z_value - 1
    else:
        return 7 - z_value

# Graficamos la funci√≥n g(z) usando matplotlib
z_values_left = np.linspace(0, 2, 400, endpoint=False)  # Rango para z < 2
z_values_right = np.linspace(2, 5, 400)  # Rango para z >= 2
g_left = [g_function(z_val) for z_val in z_values_left]
g_right = [g_function(z_val) for z_val in z_values_right]

# Crear la gr√°fica
plt.figure(figsize=(8, 6))
plt.plot(z_values_left, g_left, label=r"$3z - 1$ cuando $z < 2$", color='blue')
plt.plot(z_values_right, g_right, label=r"$7 - z$ cuando $z \geq 2$", color='green')

# Agregar puntos y anotaciones en z = 2
plt.scatter(2, g_function(2), color='green')  # L√≠mite por la derecha
plt.scatter(2, g_function(1.999), edgecolor='blue', facecolor='none', s=80)  # L√≠mite por la izquierda

# Detalles de la gr√°fica
plt.axvline(x=2, color='red', linestyle='--')
plt.title(r"Gr√°fica de la funci√≥n $g(z)$")
plt.xlabel(r"$z$")
plt.ylabel(r"$g(z)$")
plt.legend()
plt.grid(True)
plt.show()

# C√°lculo de los l√≠mites en z = 2 por la izquierda y derecha
limit_left = sp.limit(g_piecewise, z, 2, dir='-')  # L√≠mite cuando z tiende a 2 por la izquierda
limit_right = sp.limit(g_piecewise, z, 2, dir='+')  # L√≠mite cuando z tiende a 2 por la derecha
limit_value = limit_left if limit_left == limit_right else None  # Si los l√≠mites son iguales, existe el l√≠mite

# Mostrar resultados
limit_left, limit_right, limit_value

def g_izquierda(z_value_izq):
    return 3 * z_value_izq - 1

def g_derecha(z_value_der):
    return 7 - z_value_der

print("cuando el limite es 2 por la izquierda " + str(g_izquierda(2)))
print("cuando el limite es 2 por la derecha " + str(g_derecha(2)))
print("cuando los limites laterales son iguales (2) = " + str(g_function(2)))
```

```python
import numpy as np
import matplotlib.pyplot as plt


def g(t):
    return t**2 - np.cos(t)**2

t_values = np.linspace(-np.pi, np.pi, 400)
g_values = g(t_values)

plt.figure(figsize=(10, 5))
plt.plot(t_values, g_values, label='g(t) = t^2 - cos^2(t)', color='blue')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.title('Graph of g(t) = t^2 - cos^2(t)')
plt.xlabel('t')
plt.ylabel('g(t)')
plt.grid()
plt.legend()
plt.show()


num_solutions = np.sum(np.diff(np.sign(g_values)) != 0)
print(f"number to solutions of the equation t^2 = cos^2(t): {num_solutions}")


def bisection_method(func, a, b, tol=0.005):
    if func(a) * func(b) >= 0:
        raise ValueError("The function must have different signs at a and b.")

    while (b - a) / 2 > tol:
        c = (a + b) / 2.0
        if func(c) == 0:
            return c
        elif func(a) * func(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2.0


solutions = []
for i in range(len(g_values) - 1):
    if g_values[i] * g_values[i + 1] < 0:
        root = bisection_method(g, t_values[i], t_values[i + 1])
        solutions.append(root)


print("estimate solutions t^2 = cos^2(t):")
for sol in solutions:
    print(f"{sol:.5f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones de demanda y oferta
def D(p):
    return 50000 * p + 10000 * p**2

def S(p):
    return (1/3) * p**2 + 2 * p

#  Graficar las funciones de demanda y oferta
p_values = np.linspace(0, 10, 400)
D_values = D(p_values)
S_values = S(p_values)

plt.figure(figsize=(10, 6))
plt.plot(p_values, D_values, label='Demanda D(p)', color='blue')
plt.plot(p_values, S_values, label='Oferta S(p)', color='orange')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.title('Gr√°ficas de Demanda y Oferta')
plt.xlabel('Precio (p)')
plt.ylabel('Unidades')
plt.legend()
plt.grid()
plt.show()

for p in np.linspace(0, 10, 11):
    print(f"p: {p:.2f}, D(p) - S(p): {D(p) - S(p):.2f}")

# Definir la funci√≥n de bisecci√≥n para encontrar el equilibrio
def equilibrium_price(func_d, func_s, a, b, tol=0.005):
    def f(p):
        return func_d(p) - func_s(p)

    f_a = f(a)
    f_b = f(b)

    print(f"f({a}) = {f_a:.2f}, f({b}) = {f_b:.2f}")

    if f_a * f_b >= 0:
        raise ValueError("La funci√≥n debe tener signos diferentes en a y b.")

    while (b - a) / 2.0 > tol:
        c = (a + b) / 2.0
        if f(c) == 0:
            return c
        elif f_a * f(c) < 0:
            b = c
            f_b = f(c)
        else:
            a = c
            f_a = f(c)
    return (a + b) / 2.0

# Intentar encontrar el precio de equilibrio
price_equilibrium = equilibrium_price(D, S, 0, 10)
print(f"Precio de equilibrio: {price_equilibrium:.5f}")

units_at_equilibrium = D(price_equilibrium)
print(f"Unidades producidas a precio de equilibrio: {units_at_equilibrium:.5f}")

if price_equilibrium < 10:
    print("Si los productores suben el precio por encima del precio de equilibrio, habr√° un excedente.")
else:
    print("Si los productores bajan el precio por debajo del precio de equilibrio, habr√° una escasez.")

print("Si los productores aumentan la producci√≥n, pueden satisfacer la demanda creciente.")
print("Si los productores disminuyen la producci√≥n, pueden enfrentar escasez si la demanda se mantiene constante.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.sin(np.pi * t) + t

def g(t):
    return t**2

t_values = np.linspace(0, 1, 400)
f_values = f(t_values)
g_values = g(t_values)

h_values = f_values - g_values

plt.figure(figsize=(8, 6))
plt.plot(t_values, f_values, label=r'$f(t)$', color='blue')
plt.plot(t_values, g_values, label=r'$g(t)$', color='green')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

c_estimated = t_values[np.argmin(np.abs(h_values))]
plt.scatter(c_estimated, f(c_estimated), color='red', label=f'Punto c ‚âà {c_estimated:.3f} donde f(c) = g(c)')

plt.title('Demostraci√≥n gr√°fica de $f(c) = g(c)$ en [0, 1]')
plt.xlabel('t')
plt.ylabel('Valor de las funciones')
plt.legend()
plt.grid(True)
plt.show()

{
    "Punto aproximado donde f(c) = g(c)": c_estimated
}
```
